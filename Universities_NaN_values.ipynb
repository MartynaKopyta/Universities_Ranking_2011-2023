# Universities ranking data 2011-2023

```python
import pandas as pd
```

```python
df = pd.read_csv('unis_data.csv', index_col=0)
```

```python
df.isnull().sum()
```

## Managing missing values 

### Filling NaNs with a number

```python
df2 = df.fillna(value = 0)
df2
```

```python
df2.isnull().sum().sum()
```

### Filling null values with a previous value

```python
df3 = df.fillna(method = 'pad')
df3
```

```python
df3.isnull().sum().sum()
```

### Filling null values with following values

```python
df4 = df3.fillna(method = 'bfill')
df4
```

```python
df4.isnull().sum().sum()
```

Note: in order to fill NaNs with values from neighbouring columns - add parameter axis = 1

### Filling different values in null in different columns

```python
df5 = df.fillna({'scores_overall': 'No score', 'stats_female_male_ratio': 'No stats'})
df5
```

### Filling null values with the mean of a column

```python
df6 = df.fillna(value = df['stats_student_staff_ratio'].mean())
df6

#you can also fill with max, min
```

### dropna()

```python
df7 = df.dropna()
df7
```

```python
#now only rows that contain just NaNs are dropped, you can change axis to perform operation on columns

df8 = df.dropna(how='all')
df8
```

### replace()

```python
import numpy as np #you can't just type NaN
df9 = df.replace(to_replace=np.nan, value=0)
df9
```

```python
#using this function you can replace any value, not just NaN

df10 = df9.replace(to_replace = 0.0, value = 5.0)
df10
```

### interpolate()

```python
#we assume that data is linear 

df11 = df.copy()
df11['scores_teaching'] = df11['scores_teaching'].interpolate(method='linear')
df11
```

```python
#limit how many null values will be filled

df11['scores_teaching'] = df11['scores_teaching'].interpolate(method='linear', limit_direction='forward')
df11
```

## Chosen method - filling with mean or '.' considering data type

```python
df = df.apply(lambda x: x.fillna(x.mean()) if x.dtype.kind in 'biufc' else x.fillna('.'))
df
```
