# Merging csv files into one 

```python
import os
import pandas as pd
```

```python
data_file_folder = 'archive'
df =[]
for file in os.listdir(data_file_folder):
    if file.endswith('.csv'):
        print('Loading file {0}...'.format(file))
        df.append(pd.read_csv(os.path.join(data_file_folder, file)))
```

```python
len(df)
```

```python
df = pd.concat(df)
df
```

```python
df.info()
```

### Getting rid of Unnamed 21 and 25

```python
df = df.drop(df.columns[[21, 26]], axis = 1)
df
```

### Dataframe to csv for later use

```python
df.to_csv('unis_data.csv')
```
